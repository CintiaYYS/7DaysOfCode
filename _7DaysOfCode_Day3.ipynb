### Mudando tipo de dados da coluna matricula_ou_siape para string

### Adicionando coluna CDU - Classificação Decimal Universal

```python
df_dados['localizacao'].dtype
```

```python
def CDU(localizacao):
    if localizacao < 100:
      return 'Generalidades. Ciência e conhecimento'
    elif localizacao < 200:
        return 'Filosofia e psicologia'
    elif localizacao < 300:
        return 'Religião'
    elif localizacao < 400:
        return 'Ciências sociais'
    elif localizacao < 500:
        return 'Classe vaga'
    elif localizacao < 600:
        return 'Matemática e ciências naturais'
    elif localizacao < 700:
        return 'Ciências aplicadas'
    elif localizacao < 800:
        return 'Belas artes'
    elif localizacao < 900:
        return 'Linguagem.Língua.Linguística'
    elif localizacao < 1000:
        return 'Geografia.História.Ciências sociais aplicadas'
    else:
      return 'Fora do padrão'
```

```python
df_dados['CDU']=df_dados['localizacao'].apply(CDU)
```

```python
df_dados.columns
```

## Passando a coluna data_emprestimo para o tipo Datatime

```python
df_dados['data_emprestimo'] = pd.to_datetime(df_dados['data_emprestimo'], format='%Y/%m/%d %H:%M:%S.%f')
```

# Day 3 - Análise exploratória de dados

## Quantidades de exemplares emprestados por ano

```python
emprestimos_data = pd.DataFrame(df_dados['data_emprestimo'].value_counts())
```

```python
emprestimos_por_ano = emprestimos_data.groupby(emprestimos_data.index.year).sum()
emprestimos_por_ano
```

```python
emprestimos_por_ano['count'].plot(kind='line', figsize=(10, 5), title='Quantidade de empréstimos por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de empréstimos')
```

## Quantidade de exemplares emprestados por mês

```python
emprestimos_por_mes = emprestimos_data.groupby(emprestimos_data.index.month).sum()
emprestimos_por_mes
```

```python
# Nomes dos meses
nomes_meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
               'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

emprestimos_por_mes['count'].plot(kind='line', figsize=(10, 5), title='Quantidade de empréstimos por mês de 2010 a 2020')
plt.xticks(emprestimos_por_mes.index, nomes_meses, rotation=45)

plt.xlabel('Mes')
plt.ylabel('Quantidade de empréstimos')
```

Dois meses com **mais** empréstimos de exemplares: Março e Agosto
Dois meses com **menos** empréstimos de exemplares: Janeiro e Dezembro

## Quantidade de exemplares emprestados por horário

```python
emprestimos_por_hora = emprestimos_data.groupby(emprestimos_data.index.hour).sum()
emprestimos_por_hora
```

```python
sns.barplot(x=emprestimos_por_hora.index, y=emprestimos_por_hora['count'],palette='Blues',hue=emprestimos_por_hora['count'],dodge=False)

plt.xlabel('Horário')
plt.ylabel('Quantidade de empréstimos')
plt.title('Quantidade de empréstimos por horário')
```

Horário com maior movimento de empréstimos: 10h e 16h
Horário com menor movimento de empréstimos: 7h e 21h
